# You just have to add more cols with some extent of powers of features

```python
from sklearn.pipeline import Pipeline as pipe
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
x=6*np.random.rand(300,1)-3
y=2*x**2-4*x+2+np.random.randn(300,1)
plt.scatter(x,y)
plt.show()
```

```python
poly=PolynomialFeatures(degree=2,include_bias=False) # will not iclude x^0
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
x_train_after=poly.fit_transform(x_train)
x_test_after=poly.transform(x_test)
model=LinearRegression()
model.fit(x_train_after,y_train)
res=model.predict(x_test_after)
plt.plot(x_train,y_train,'b.')
plt.plot(x_train,model.predict(x_train_after),'r.')
print(r2_score(res,y_test))
```
