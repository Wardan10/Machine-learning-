# Feature importance

For identifying the important features with the help of decision trees and random forrest too, we can classify the columns

```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
x,y=load_digits(return_X_y=True)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)
plt.imshow(x[39].reshape(8,8))
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
import numpy as np
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(x_train,y_train)
print(model.predict(x[39].reshape(1,-1)))
model.feature_importances_
plt.imshow(model.feature_importances_.reshape(8,8))
plt.show()
```
