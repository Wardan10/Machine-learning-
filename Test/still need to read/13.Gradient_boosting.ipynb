# Gradient Boosting:

Ensemble technique.

In regression , the first model is just finding the average , then the next models will be 

Finding the residual :( actual - predicted )

all the next models will be decisions trees whose inpits will be features and the outputs will be the residuals whose value to be added to the first model to make closer to actual. 

Now:
prediction = M1 + lr * M2

then we again calculate the residual and again predict the value to be added.

prediction = M1 + lr* M2  + lr*M3
