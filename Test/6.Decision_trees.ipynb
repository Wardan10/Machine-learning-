# Entropy:

Entropy is measure of disorder.More knowledge less entropy,Then how to calculate it?? 

**E= all the values of (- p_i * log2( p_i ))**

**Gini Impurity:**

GI=( 1- sum of all the (P_i^2) )

# We can choose the model according to entropy or Gini impurity.

**Information Gain:**

This metric measures the quality of a split.Constructing descision tree is all about finding the attribute that returns the highes information gain.

IG= E(parent) - {weighted avg} * E{children}

->We select the column according to spitting wrt particular coloumn and select the one which have the largest IG.

for numerical data , we split the data into 2 or more cols according all the possible values and select the one with maximum information gain.
